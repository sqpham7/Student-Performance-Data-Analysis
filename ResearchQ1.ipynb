## Random Forests Tuning

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

depths=[3,4,5,6,7]
estimators=[50,100,150,200,250]

best_mean_score = 0
best_params = {'n_estimators': None, 'max_depth': None}

kf =KFold(n_splits=5, shuffle=True, random_state=0)


for estimator in estimators:
    for depth in depths:
        clf = RandomForestClassifier(n_estimators=estimator, verbose=0, bootstrap=True, max_depth=depth, random_state=0) 
        CVscores= cross_val_score(clf, X_train, Y_train, scoring='accuracy', cv=kf)
        print('With n_estimators={} and with max_depth={}, the Cross validation scores mean is {}:'.format(estimator,depth,CVscores.mean()))
        
        #Check if the current combination has a higher mean score

        if CVscores.mean() > best_mean_score:
            best_mean_score = CVscores.mean()
            best_params['n_estimators'] = estimator
            best_params['max_depth'] = depth

print(f"\nBest parameters: {best_params}")
print(f"Best mean score: {best_mean_score}")
```
